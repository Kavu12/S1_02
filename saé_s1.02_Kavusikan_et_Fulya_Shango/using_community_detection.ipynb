```python
list_of_friends = ["Alice", "Bob","Dominique","Alice","Bob","Charlie"]

dico = {
  "Alice" : ["Bob", "Dominique"],
  "Bob" : ["Alice", "Charlie", "Dominique"],
  "Charlie" : ["Bob"],
  "Dominique" : ["Alice", "Bob"]
}

# Question 1 #

def create_network(list_of_friends):
    dico = {}
    i = 0
    while i < len(list_of_friends):
        if i % 2 == 0 :
            if list_of_friends[i] in dico:
                dico[list_of_friends[i]].append(list_of_friends[i+1])
            elif list_of_friends[i] not in dico:
                dico[list_of_friends[i]] = [list_of_friends[i+1]]
        elif i % 2 == 1:
            if list_of_friends[i] in dico:
                dico[list_of_friends[i]].append(list_of_friends[i-1])
            elif list_of_friends[i] not in dico:
                dico[list_of_friends[i]] = [list_of_friends[i-1]]
        i += 1
    return dico




# Question 3 #

def get_people(network):
    liste = []
    i = 0
    while i < len(network):
        if network[i] not in liste :
            liste.append(network[i])
        i += 1
    return liste


# Question 4 #

def are_friends(network, person1, person2):
    if person2 in network[person1]:
        return False
    return True





# Question 5 #

def all_his_friend(network, person, group):
    for i in range(len(group)):
        if group[i] not in network[person]:
            return False
    return True




# Question 6 #

def is_a_community(network, group):
    for l in range(len(group)):
        for i in range(len(group)):
            if group[i] not in network[group[l]] and group[l] != group[i]:
                return False
    return True




# Question 7 #
"Paramètre: dico, groupe de personne" \
"Return: un tableau de personne qui sont amis avec tous les membres du dico"
def find_community(network, group):
    community = []
    t = []
    for i in range(len(group)):
        t.append(group[i])

        if is_a_community(network,t):
            community.append(group[i]) #Ajout de la personne dans le tableau

        elif not is_a_community(network, t):
            t.pop()

    return community




# Question 8 #
"Paramètre: dico, groupe de personne" \
"Return: un tableau qui contient les personnes qui ont le plus d'amis à ce qui ont le moins. "
def order_by_decreasing_popularity(network):
    for i in range(len(group)-1):
        for l in range(len(group)-1):
            if len(network[group[l]])< len(network[group[l+1]]):
                group[l], group[l+1] = group[l+1], group[l]
    return group




# Question 9 #
"Paramètre: dico" \
"Return: un tableau contenant ceux qui ont plus d'un ami"
def find_community_by_decreasing_popularity(network):
    t = get_people(network)
    popularity = order_by_decreasing_popularity(network, t)
    return find_community(network,popularity)




# Question 10 #
"Paramètre: dico, une personne du dico" \
"Return: un tableau contenant la personne choisi en paramètre et ses amis si ils sont dans le dico"
def find_community_from_person(network, person):
    community = [person]
    popularity = order_by_decreasing_popularity(network,network[person])

    for i in range(len(popularity)):
        if all_his_friends(network,popularity[i],community):
            community.append(popularity[i])

    return community




# Question 12#
"Paramètre: dico" \
"Return: un tableau qui est la plus grande communauté du dico"
def find_max_community(network):
    person = get_people(network)
    maximum = find_community_from_person(network, pers[0])
    for i in range(len(person)):
        if len(find_community_from_person(network,person[i]))> len(maximum):
            maximum = find_community_from_person(network, person[i])

    return maximum

```

# Question 2 :

`create_network` a une complexité en 8n + 3, une complexité linéaire, en O(n).
    
`dico_reseau` a une complexité en  36n² + 13n + 5, une complexité quadratique en O(n²).

```python
from time import time

def time_create_network():
    tic = time()
    
    create_network(list_of_friends)
    
    tac = time()
    
    print(round(1000*(tac-tic),2),"ms")
    
    
time_create_network()

def liste_amis(amis, prenom):
    """
        Retourne la liste des amis de prenom en fonction du tableau amis.
    """
    prenoms_amis = []
    i = 0
    while i < len(amis)//2:
        if amis[2 * i] == prenom:
            prenoms_amis.append(amis[2*i+1])
        elif amis[2*i+1] == prenom:
            prenoms_amis.append(amis[2*i])
        i += 1
    return prenoms_amis


def personnes_reseau(amis):
    """ Retourne un tableau contenant la liste des personnes du réseau."""
    people = []
    i = 0
    while i < len(amis):
        if amis[i] not in people:
            people.append(amis[i])
        i += 1
    return people

def dico_reseau(tab):
    """ Retourne le dictionnaire correspondant au réseau."""
    dico = {}
    people = personnes_reseau(tab)
    i = 0
    while i < len(people):
        dico[people[i]] = liste_amis(tab, people[i])
        i += 1
    return dico


def time_dico_reseau():
    tic = time()
    
    dico_reseau(list_of_friends)
    
    tac = time()
    
    print(round(1000*(tac-tic),2),"ms")
    
time_dico_reseau()
```
